```python
import matplotlib.pyplot as plt
import numpy as np

# Model names and metrics
models = ["Naive", "Linear Reg.", "Random Forest", "LSTM"]
accuracy = [50.00, 56.20, 65.40, 72.50]
rmse = [6.89, 5.24, 4.12, 3.89]

# X-axis positions
x = np.arange(len(models))
width = 0.35

# Create plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar plot for accuracy
bar1 = ax1.bar(x - width/2, accuracy, width, label='Accuracy (%)', color='skyblue')
ax1.set_ylabel('Directional Accuracy (%)', color='blue')
ax1.set_ylim(0, 100)
ax1.set_xticks(x)
ax1.set_xticklabels(models)
ax1.set_title('Model Comparison: Directional Accuracy vs RMSE')

# Secondary Y-axis for RMSE
ax2 = ax1.twinx()
bar2 = ax2.bar(x + width/2, rmse, width, label='RMSE', color='salmon')
ax2.set_ylabel('RMSE (Close Price)', color='red')
ax2.set_ylim(0, 8)

# Annotate bars
for i in range(len(models)):
    ax1.text(x[i] - width/2, accuracy[i] + 1, f"{accuracy[i]:.1f}%", ha='center', color='blue', fontsize=9)
    ax2.text(x[i] + width/2, rmse[i] + 0.2, f"{rmse[i]:.2f}", ha='center', color='red', fontsize=9)

# Legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Model names and their directional accuracy
models = ["Naive", "Linear Reg.", "Random Forest", "LSTM"]
accuracy = [50.00, 56.20, 65.40, 72.50]

# Colors for the pie chart
colors = ['lightgrey', 'lightblue', 'orange', 'green']

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    accuracy,
    labels=models,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    shadow=True,
    explode=(0, 0.05, 0.1, 0.15)
)

plt.title('Model Accuracy Distribution')
plt.tight_layout()
plt.show()
```
